```python
import matplotlib.pyplot as plt

# Given parameters
p = 0.74
q = 1 - p

# Calculate probabilities until they are less than 0.005
k_values = []
probabilities = []
k = 1
while True:
    prob = p * (q ** (k - 1))
    if prob < 0.005:
        break
    k_values.append(k)
    probabilities.append(prob)
    k += 1

# Plotting
plt.bar(k_values, probabilities, color='c')
plt.axvline(x=1.35135, color='m', linestyle='-.', label='Expectation')
plt.axvline(x=1 , color='g', linestyle='-.', label='Median')
plt.xlabel('Number of Years')
plt.ylabel('Probability')
plt.title('Geometric Distribution with p=0.74')
plt.legend()
plt.show()
```
