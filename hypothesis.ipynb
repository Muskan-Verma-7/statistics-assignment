```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Given data
old_mean = 812
old_std = 54.6
old_variance = old_std**2
new_weights = [821, 818, 777, 732, 790, 810, 868, 845, 810, 824]
new_variance = np.var(new_weights, ddof=1)  # Sample variance

# F-test
F_statistic = new_variance / old_variance
df1 = len(new_weights) - 1  # degrees of freedom for new system
df2 = 1000 - 1  # degrees of freedom for old system (large sample approximation)
alpha = 0.05
F_critical = stats.f.ppf(1-alpha, df1, df2)

# Visualization
x = np.linspace(650, 950, 1000)
y_old = stats.norm.pdf(x, old_mean, old_std)
y_new = stats.norm.pdf(x, np.mean(new_weights), np.std(new_weights, ddof=1))

plt.figure(figsize=(10,6))
plt.plot(x, y_old, label='Old System', color='blue')
plt.plot(x, y_new, label='New System Sample', color='red')
plt.title('Distribution of Hammer Weights')
plt.xlabel('Weight (g)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

F_statistic, F_critical
```
